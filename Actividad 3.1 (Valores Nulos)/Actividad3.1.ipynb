```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ventas_totales.csv',index_col=0)
df
```

**Aqui se aplico un index_col = 0**
Esto se aplico para eliminar la columna predeterminada que se genera automaticamente del lado derecho

```python
valores_nulos =df.isnull().sum()
valores_nulos
```

**Se aplico una especie de filtro para los nulos**
Con este filtro se muestra cuantos nulos hay en total en cada columna de este DataFrame

```python
valores_nulos =df.isnull().sum().sum()
valores_nulos
```

**Se sumaron y dieron el total de nulos**
Con esto se muestra el total de nulos que hay por columna en total del dataframe

```python
columnas_sin_nulos=df.copy()
```

**Se realizo una copia**
Realice la copia pero con ella solo vamos a mostrar todas las columnas que no tienen nulos en ellas

```python
columnas_sin_nulos["ventas_precios_corrientes"] = columnas_sin_nulos["ventas_precios_corrientes"].fillna(round(df["ventas_precios_corrientes"].mean(),1))
```

**Se aplico la eliminarcion en esta columna**
Se realizo esto para demostrar que anterior mente esta columna no manejaba ningun valor nulo entonces para no realizar lo mismo con las demas columnas, no se va aplicar en las demas tablas que no tengan igualmente valores nulos

```python
valores_nulos =columnas_sin_nulos.isnull().sum()
valores_nulos
```
